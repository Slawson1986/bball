```python
import os
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.formula.api as sm
from sklearn.cluster import KMeans
# custom modules
import data_getters, model, boxscores, optimizer
from entities import Player, Game
from pprint import pprint
```

```python
# get data, create useful structures for later
latest_day = datetime.date(2014,4,17)

player_data, game_data = boxscores.load_jsons_to_memory('boxscores_advanced/2013', latest_day)
players = [p.get_name() for p in player_data]
players_stats = {} #k,v -> player name (str), player data (dict)

teams = {}
team_names = set() 
for player in players:
    data = player_data[players.index(player)]
    team_abbr = data.adv_info['TEAM_ABBREVIATION'][0]
    team_names.add(team_abbr)
    if team_abbr in teams.keys():
        teams[team_abbr].append(player)
    else:
        teams[team_abbr] = [player]
        
    players_stats[player] = data
```

```python
# get all the games that a team has played in 
def get_team_games(team_abbr):
    games = []
    for game in game_data:
        if game.away_info['TEAM_ABBREVIATION'] == team_abbr or game.home_info['TEAM_ABBREVIATION'] == team_abbr:
            games.append(game)
    return games
```

```python
# CLUSTERING PLAYER PERFORMANCE MODEL API
# Runs k-means clustering on historical player performance against a particular team's defense.
# Calculates the percentage of the time that a player is in a particular cluster
# player performance =  sum_x of (% of time in cluster x)*(cluster centroid)

from math import pow

# returns two dictionarys, with keys = cluster labels, values = data point array (or data point array w/ player name)
def get_points_in_cluster(km, data_points, data_points_labeled):
    clusters = {}
    clusters_labeled = {}
    player_labels = {} 
    for b in range(len(np.unique(km.labels_))):
        clusters[b] = []
        clusters_labeled[b] = [] 
    i = 0
    for label in km.labels_:
        clusters_labeled[label].append(data_points_labeled[i])
        clusters[label].append(data_points[i])
        player_labels[data_points_labeled[i][0]] = label
        i+=1
    return clusters, clusters_labeled 

def dist(x,y):
    dist = 0.0
    for i in range(len(x)):
        dist += pow(abs(x[i] - y[i]),2)
    return dist
    
def inter_cluster_dist(centroid1, centroid2):
    dist = 0.0
    for i in range(len(centroid1)):
        dist += pow(abs(centroid1[i] - centroid2[i]),2)
    return dist
    
#cluster is an array of the points in a cluster
def intra_cluster_dist(cluster):
    max_dist = 0.0
    for i in range(len(cluster)):
        for j in range(len(cluster)):
            d = dist(cluster[i], cluster[j])
            if d > max_dist:
                max_dist = d
    return max_dist

# since we use sklearn, just pass in the resulting clustering object for ease
def dunn_index(km, clusters):
    intra_cluster_distances = []
    for key in clusters.keys():
        intra_cluster_distances.append(intra_cluster_dist(clusters[key]))
    
    d_prime_k  = float(np.max(intra_cluster_distances))

    dunn_index = float('Inf')
    for i in range(len(km.cluster_centers_)):
        for j in range(len(km.cluster_centers_)):
            if i != j:
                new_index = float(inter_cluster_dist(km.cluster_centers_[i], km.cluster_centers_[j]) / d_prime_k )
                #new_index = float(d_prime_k /inter_cluster_dist(km.cluster_centers_[i], km.cluster_centers_[j]))
                if new_index < dunn_index:
                    dunn_index = new_index
    return dunn_index
                
# get data for k-means clustering
def get_km_data(defense_team_abbr, desired_stats=['PTS', 'OREB', 'DREB', 'STL', 'AST', 'BLK']):
    all_games = []
    games = get_team_games(defense_team_abbr)

    # this is stupid, was trying a diff idea, not really needed
    for game in games:
        all_games.append(game)


    data_points = [] # points for clustering
    data_points_labeled = [] # points with labels

    for game in all_games:
        if game.away_info['TEAM_ABBREVIATION'] == defense_team_abbr:
            off_team_abbr = game.home_info['TEAM_ABBREVIATION'] 
        else:
            off_team_abbr = game.away_info['TEAM_ABBREVIATION']
        game_id = game.home_info['GAME_ID'] #game id
        for player in teams[off_team_abbr]:
            try:
                index = players_stats[player].info['GAME_ID'].index(game_id)
                features = []
                for stat in desired_stats:
                    stat_val = players_stats[player].info[stat][index]
                    if stat_val == None:
                        break
                    features.append(stat_val)
                if len(features) > 0:
                    data_points.append(features)
                    data_points_labeled.append((player,features,game_id))

            except Exception as e:
               # print e.message
                continue
    return data_points, data_points_labeled

# get percentage of time a player is in a cluster
def player_cluster_mixture(clusters_labeled):
    player_percentages = {}
    
    # first get data into player_name, [cluster_labels] format
    for key in clusters_labeled.keys():
        for i in range(len(clusters_labeled[key])):
            if clusters_labeled[key][i][0] not in player_percentages:
                player_percentages[clusters_labeled[key][i][0]] = [key]
            else:
                player_percentages[clusters_labeled[key][i][0]].append(key)
    # now calculate what percentage of the time each player is in a cluster
    
    for key in player_percentages.keys():
        percentages = []
        for i in range(len(clusters_labeled.keys())):
            percentages.append(float(player_percentages[key].count(i))/float(len(player_percentages[key])))
        player_percentages[key] = percentages
    return player_percentages

# predict player performance using clustering 
def cluster_model_predict(km, player_percentages):
    projections = {} 
    for player in player_percentages.keys():
        #print player_percentages[player]
        cluster_performances = [np.multiply(np.array(km.cluster_centers_[i]), player_percentages[player][i]) for i in range(len(player_percentages[player]))]
        #print km.cluster_centers_
        total = np.zeros(len(cluster_performances[0]))
        for i in range(len(cluster_performances)):
            total = np.add(total, cluster_performances[i])
        projections[player] = total
    return projections
        
```
