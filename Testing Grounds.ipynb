```python
# calculates expected fantasy points
# DK Formula:
#   Point = +1 PT
#   Made 3pt. shot = +0.5 PTs
#   Rebound = +1.25 PTs
#   Assist = +1.5 PTs
#   Steal = +2 PTs
#   Block = +2 PTs
#   Turnover = -0.5 PTs
#   Double-Double = +1.5PTs (MAX 1 PER PLAYER: Points, Rebounds, Assists, Blocks, Steals)
#   Triple-Double = +3PTs (MAX 1 PER PLAYER: Points, Rebounds, Assists, Blocks, Steals)
def expected_fp(player):
    fp = 0
    doubled = 0
    weights = {'PTS': 1.0, '3PTM': 0.5, 'REB': 1.25, 'AST': 1.5, 'STL': 2.0, 'BLK': 2.0, 'TO': -0.5}
    for s, w in weights:
        v = player.predict(s)
        fp += v * w
        if s in ['PTS', 'REB', 'AST', 'BLK', 'STL']:
            if v >= 10.0:
                doubled += 1
    if doubled == 2:
        fp += 1.5
    elif doubled >= 3:
        fp += 4.5 # assume that you get both DD and TD bonus
    return fp

# optimizes expected FP based on total salary and positional constraints
# players: list of Player objects
# total_salary: a float representing total salary that can be spent
# assumes roster positions=['PG','SG','SF','PF','C','G','F','UTIL']
# returns a subset of players which maximizes expected FP given salary constraints

# maximize sum of p.DKppg under the constraints:
#   sum of p.salary <= total_salary
#   1 <= n(PG) <= 2
#   1 <= n(SG) <= 2
#   1 <= n(SF) <= 2
#   1 <= n(PF) <= 2
#   1 <= n(C) <= 2
#   4 <= n(PG) + n(SG) + n(C) <= 5
#   4 <= n(SF) + n(PF) + n(C) <= 5
#   n(PG) + n(SG) + n(SF) + n(PF) + n(C) = 8

def objective(x, players):
    return -x.dot(np.array([p.DKppg for p in players]))

def objective2(x, players, total_salary):
    s = -x.dot(np.array([p.DKppg for p in players]))
    
    pg = np.array([1 if p.position == 'PG' else 0 for p in players])
    sg = np.array([1 if p.position == 'SG' else 0 for p in players])
    sf = np.array([1 if p.position == 'SF' else 0 for p in players])
    pf = np.array([1 if p.position == 'PF' else 0 for p in players])
    c = np.array([1 if p.position == 'C' else 0 for p in players])
    
    n_pg = x.dot(pg)
    n_sg = x.dot(sg)
    n_sf = x.dot(sf)
    n_pf = x.dot(pf)
    n_c = x.dot(c)
    
    penalty = -100000

    if sum(x) != 8:
        return penalty
    if x.dot(np.array([p.salary for p in players])) > total_salary:
        return penalty
    if n_pg < 1 or n_pg > 2:
        return penalty
    if n_sg < 1 or n_sg > 2:
        return penalty
    if n_sf < 1 or n_sf > 2:
        return penalty
    if n_pf < 1 or n_pf > 2:
        return penalty
    if n_c < 1 or n_c > 2:
        return penalty
    if n_pg + n_sg + n_c < 4 or n_pg + n_sg + n_c > 5:
        return penalty
    if n_sf + n_pf + n_c < 4 or n_sf + n_pf + n_c > 5:
        return penalty
    
    return s
    
def optimize(players, total_salary=50000):
    """TODO"""
    ind = np.zeros(len(players))
    ind[0] = 1
    ind[1] = 1
    #pg = np.array([1 if p.position == 'PG' else 0 for p in players])
    #sg = np.array([1 if p.position == 'SG' else 0 for p in players])
    #sf = np.array([1 if p.position == 'SF' else 0 for p in players])
    #pf = np.array([1 if p.position == 'PF' else 0 for p in players])
    #c = np.array([1 if p.position == 'C' else 0 for p in players])
    
    #print minimize(objective2, ind, args=(players,total_salary))

    return

# imports a file from draftkings containing salary info
# returns a list of Player objects
def import_draftkings_salary(filename):
    players = []
    with open(filename, 'rb') as f:
        csvreader = csv.reader(f, delimiter=',')
        col_names = csvreader.next()
        for row in csvreader:
            p = Player(row[1], row[0], float(row[2]), row[3], float(row[4]))
            players.append(p)
    return players

fn = 'draftkings_data/DKSalaries_11042014.csv'
players = import_draftkings_salary(fn)
#optimize(players)
```
